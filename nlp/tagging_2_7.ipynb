```python
import json
from treetagger import TreeTagger
tt = TreeTagger(language='russian',encoding='latin-1')
```

```python
with open("../output/enhanced_dataset.json","r") as fp:
    db = json.load(fp)
```

```python
test = u"КРАСНОПОЛЬ , местечко Могилевского Наместничества в Чериковском уезде. Здесь бывают четыре ярмарки в год.".encode('utf-8')
print(len([c for c in test]))
print(len(test))
print(test)
print(type(test))
```

```python
tt.tag(u"КРАСНОПОЛЬ , местечко Могилевского Наместничества в Чериковском уезде. Здесь бывают четыре ярмарки в год.".encode('utf-8'))
```

```python
for k,v in db.iteritems():
    print(v['description_new_orth'])
    text = v['description_new_orth'].encode('utf-8')
    db[k]['desc_tags'] = tt.tag(text)
```

```python
for k,v in db.iteritems():
    print(v['desc_tags'])
```

```python
with open('../output/enhanced_dataset_tagged.json',"w") as fp:
    json.dump(db,fp)
```
