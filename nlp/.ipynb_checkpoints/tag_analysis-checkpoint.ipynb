```python
import json
with open("../resources/multext.json","r") as fp:
    main_lookup = json.load(fp)
def pos_lookup(code):
    out_dict = {}
    try:
        lookup = main_lookup[code[0]]
    except KeyError:
        return {"CODE":code}
    if code == "SENT":
        return {"CODE":code}
    for i in range(0,len(code)):
        if code[i]!='-':
            try:
                out_dict[lookup[i][0]]=lookup[i][1][code[i]]
            except KeyError:
                out_dict[lookup[i][0]]="*{}".format(code[i])
                #print("{} not found in {}, type = {}".format(code[i],lookup[i],code[0]))
    return out_dict
```

```python
with open('../output/enhanced_dataset_tagged.json','r',encoding='utf-8') as fp:
    db = json.load(fp)
```

```python
for k,v in db.items():
    db[k]['ex_tags'] = []
    for tag in v['desc_tags']:
        temp = {"ORIGINAL_WORD":tag[0],"LEMMA":tag[2]}
        temp.update(pos_lookup(tag[1]))
        db[k]['ex_tags'].append(temp)
```

```python
for k,v in db.items():
    db[k]['sent_tags']=[]
    temp = []
    for w in v['desc_tags']:
        temp.append(w)
        if w[1] == 'SENT':
            db[k]['sent_tags'].append(temp)
            temp = []
```

```python
for k,v in db.items():
    db[k]['ex_sent_tags']=[]
    temp = []
    for w in v['ex_tags']:
        temp.append(w)
        if 'CODE' in w and w['CODE'] == 'SENT':
            db[k]['ex_sent_tags'].append(temp)
            temp = []
```

```python
def agreement_check(dict1,dict2,conds):
    for c in conds:
        try:
            if dict1[c]==dict2[c]:
                continue
            else:
                return False
        except KeyError:
            return False
    return True
```

```python
def case_finder(desc_tags,case_pattern,wordlist):
    a,b,c = zip(*desc_tags)
    if any(t in c for t in wordlist):
        outlist = []
        for w in desc_tags:
            if re.match(case_pattern,w[1]):
                if w[2]!='<unknown>':
                    outlist.append(w[2])
                else:
                    outlist.append(w[0])
        return outlist
```
