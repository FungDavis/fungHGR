```python
import json
```

```python
with open("holidays.json","r",encoding='utf-8') as fp:
    holidays = json.load(fp)
```

```python
new_holidays = {0:{'names':None,'easter_offset':None}}
extant = []
for k in holidays:
    if holidays[k] in extant:
        for h in new_holidays:
            if new_holidays[h]['easter_offset']==holidays[k]:
                new_holidays[h]['names'].append(k)
    else:
        extant.append(holidays[k])
        new_holidays[max(new_holidays.keys())+1] = {'names':[k],'easter_offset':holidays[k]}
```
